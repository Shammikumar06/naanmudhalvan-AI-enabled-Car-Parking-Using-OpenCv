```python
import statsmodels.api as sm

#define response variable
y = dataset['Price']

#define explanatory variable
x = dataset[['lot_area_renov']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```

Multi-Variate Analysis

```python
import pandas as pd
import seaborn as sns

# Load the dataset
dataset = pd.read_csv('House Price India.csv')

# Correlation matrix
corr_matrix = dataset.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

# Scatter plot
sns.scatterplot(x='lot_area_renov', y='Price', data=dataset)
plt.title('Scatter plot of Price vs lot_area_renov')
plt.show()

# Pairplot
sns.pairplot(dataset, vars=['id', 'Date', 'lot_area_renov', 'Lattitude', 'Longitude', 'Price'])
plt.show()

# Regression plot
sns.regplot(x='lot_area_renov', y='Price', data=dataset)
plt.title('Regression plot of Price vs lot_area_renov')
plt.show()
```
