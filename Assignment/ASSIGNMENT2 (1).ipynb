```python
dataset = pd.read_csv('House Price India.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().any()
```

label encoding*

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Price'] = le.fit_transform(dataset['Price'])
```

```python
dataset.head()
```

```python
x  = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:5].values
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z=one.fit_transform(x[:,3:4]).toarray()
x=np.delete(x,3,axis=1)
x=np.concatenate((z,x),axis=1)
```

```python
x
```

train test and split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

INPUT LAYER

```python
model = Sequential()
```

```python
x_train.shape
```

```python
model.add(Dense(units = 4,activation = 'relu'))
```

HIDDEN LAYER

```python
model.add(Dense(units = 12,activation = 'relu')) # 1st hidden layer
```

```python
model.add(Dense(units = 8,activation = 'relu')) # second hiden layer
```

```python
model.add(Dense(units = 9,activation = 'relu')) # third Hidden layer
```

OUTPUT LAYER

```python
model.add(Dense(units = 1))
```

COMPILE THE MODEL

```python

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train , batch_size = 15,epochs = 500)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
```
